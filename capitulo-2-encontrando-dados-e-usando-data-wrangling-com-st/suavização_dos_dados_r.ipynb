```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

path_ = "C:\\Users\\mateu\\Documents\\MEGA\\Projetos-git\\analise-pratica-series-temporais_predicao-stats-ml\\capitulo-2-encontrando-dados-e-usando-data-wrangling-com-st\\dataset\\passengers\\AirPassengers.csv"
# Lendo arquivo 
passer = pd.read_csv(path_)
```

```python
passer.head(10)
```

```python
plt.plot(
    passer['date'], passer['passengers']
)
```

```python
# decaimento exponencial 

passer['smooth.5'] = passer['passengers'].ewm(alpha=0.5).mean()
passer['smooth.9'] = passer['passengers'].ewm(alpha=0.9).mean()
```

```python
passer.head(10)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.plot(passer['passengers'])
plt.subplot(1,3,2)
plt.plot(passer['smooth.5'])
plt.subplot(1,3,3)

plt.plot(passer['smooth.9'])
plt.show()
```

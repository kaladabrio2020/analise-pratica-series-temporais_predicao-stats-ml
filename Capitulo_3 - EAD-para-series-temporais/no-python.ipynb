```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Exemplo de uma série temporal sintética
np.random.seed(42)
data = np.random.randn(100)  # Dados aleatórios para exemplo

# Plotar ACF
plt.figure(figsize=(10,5))
plot_acf(data, lags=20)
plt.show()

# Plotar PACF
plt.figure(figsize=(10,5))
plot_pacf(data, lags=20)
plt.show()
```

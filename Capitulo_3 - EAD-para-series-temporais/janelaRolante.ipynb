```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = abs(np.random.RandomState(42).randn(100) * 100)
x = pd.Series(x, name="series")
```

```python
plt.plot(x)
```

```python
plt.plot(x)
plt.plot(x.rolling(window=10).mean())
x.rolling(window=10).min().plot()
```

```python
lags_ = [ x.autocorr(lag=i) for i in range(1, 20) ]
plt.plot(lags_, 'o-')
```

```python
x.shift(2)
```

```python
x.shift(1).corr(x)
```

```r
library(zoo)
library(data.table)
```

```r
x = 1:100

y = sin(x*pi/3)

par(bg = 'white')
options(repr.plot.width = 15, repr.plot.height = 10)
plot(y, type='b')
acf(y)
```

```r
cor(
    y, shift(y, 1), use = 'pairwise.complete.obs'
)
```

```r
cor(
    y, shift(y, 2), use = 'pairwise.complete.obs'
)
```

```r
x = 1:100

y = sin(x*pi/3)

par(bg = 'white')
options(repr.plot.width = 15, repr.plot.height = 10)
plot(y[1:30], type='b')
pacf(y)
```
